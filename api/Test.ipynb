```python
import requests
import pandas as pd

# Test simple
response = requests.post(
    "http://localhost:8000/api/demand/forecast",
    json={"zone_id": 3, "datetime": "2026-06-06T14:30:00"}
)
print(response.json())

# Tester plusieurs zones
results = []
for zone in [1, 3, 5, 8, 12]:
    resp = requests.post(
        "http://localhost:8000/api/demand/forecast",
        json={"zone_id": zone, "datetime": "2026-06-06T18:00:00"}
    )
    results.append(resp.json())

# Afficher dans un DataFrame
df = pd.DataFrame(results)
print(df)
```
